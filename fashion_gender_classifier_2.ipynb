```python
genders = []
features = []
decades = []
orig_indices = []

subset = df_clean.dropna(subset=["gender_dep_pron"])

for row in subset.itertuples():
    if row.gender_dep_pron[0] in ("t", "z", "n", "o"):
        continue
    adjs = [adj for adj in row.adjectives_fashion if adj is not None]

    if len(adjs) == 0:
        continue

    adjs_2 = [adj for adj in row.adjectives_char if adj is not None]


    # append consistently
    genders.append(row.gender_dep_pron)
    # features.append([f"term_{row.term.lower()}"])
    features.append(
    [f"term_{row.term.lower()}"] +
    [f"adj_f_{adj.lower()}" for adj in adjs] + [f"adj_c_{adj.lower()}" for adj in adjs_2]
)

    decades.append(row.decade)
    orig_indices.append(row.Index)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score




encoder = CountVectorizer(
    input="content",
    lowercase=False,
    preprocessor=None,
    tokenizer=lambda x: x,
    min_df=20,
)
encoder.fit(features)

le = LabelEncoder()
le.fit(genders)


def get_features(split):
    split_g = []
    split_f = []
    for i in split:
        split_g.append(genders[i])
        split_f.append(features[i])
    return split_g, split_f


def train_classifier(split):
    genders, features = get_features(split)
    X = encoder.transform(features)
    y = le.transform(genders)
    model = LogisticRegression(penalty="l1", solver="liblinear")
    model.fit(X, y)

    print(f1_score(y[:1000], model.predict(X[:1000])))

    # print(model.score(X[:1000], y[:1000]))
    return model
```

```python
import pandas as pd

def get_error_dataframe_with_metadata(model, split, sentence_metadata):
    """
    Build a DataFrame of prediction errors with optional metadata.

    Parameters
    ----------
    model : trained classifier
    split : list/array of indices into features array (0..len(features)-1)
    sentence_metadata : pd.Series of strings (indexed by the original dataset)

    Returns
    -------
    pd.DataFrame with columns:
        - index_in_split
        - index_in_dataset
        - true_label
        - predicted_label
        - error_type
        - tokens
        - num_tokens
        - metadata
    """

    # Get features and labels for the selected rows
    genders_split, features_split = get_features(split)

    # Vectorize
    X = encoder.transform(features_split)
    y_true = le.transform(genders_split)
    y_pred = model.predict(X)

    # Decode labels
    true_labels = le.inverse_transform(y_true)
    pred_labels = le.inverse_transform(y_pred)

    rows = []

    for row_idx, (t_true, t_pred) in enumerate(zip(true_labels, pred_labels)):
        if t_true != t_pred:
            # original_index = split[row_idx]  # index in the original dataset
            original_index = orig_indices[split[row_idx]]  # map back to original df
            # row = sentence_metadata[original_index]
            row = sentence_metadata.loc[original_index]

            print(type(sentence_metadata.loc[original_index, 'adjectives_fashion']))
            print(sentence_metadata.loc[original_index, 'adjectives_fashion'])



            rows.append({
                "index_in_split": split[row_idx],
                "index_in_dataset": original_index,
                "true_label": t_true,
                "predicted_label": t_pred,
                "error_type": (
                    "False Positive" if t_pred != t_true and t_pred == le.classes_[1]
                    else "False Negative"
                ),
                "tokens": features_split[row_idx],
                "num_tokens": len(features_split[row_idx]),
                # Safely get the sentence string from Series
                # "metadata": sentence_metadata.iloc[original_index][0],
                # "term": sentence_metadata.iloc[original_index][1],
                # "adj_fashion": sentence_metadata.iloc[original_index][2],
                # "adj_char": sentence_metadata.iloc[original_index][3],
                # "metadata": row["sentence"],
                # "term": row["term"],
                "adj_fashion": row["adjectives_fashion"],
                "adj_char": row["adjectives_char"],
#                 "tokens": features_split[row_idx],
#                 "num_tokens": len(features_split[row_idx]),
#                 "metadata": sentence_metadata.loc[original_index]
                "metadata": sentence_metadata.loc[original_index, "sentence"],


                "term": sentence_metadata.loc[original_index, "term"],
                "adj_fashion": sentence_metadata.loc[original_index, "adjectives_fashion"],
                "adj_character": sentence_metadata.loc[original_index, "adjectives_char"]





            })

    df_rows = pd.DataFrame(rows)
    return df_rows
```

```python
# Access each field safely
def safe_get_value(df, index, column):
    value = df.loc[index, column]
    if isinstance(value, pd.Series):
        return value.iloc[0]
    return value




def get_error_dataframe_with_metadata(model, split, sentence_metadata):

    # Get features and labels for the selected rows
    genders_split, features_split = get_features(split)

    # Vectorize
    X = encoder.transform(features_split)
    y_true = le.transform(genders_split)
    y_pred = model.predict(X)

    # Decode labels
    true_labels = le.inverse_transform(y_true)
    pred_labels = le.inverse_transform(y_pred)

    rows = []




    for row_idx, (t_true, t_pred) in enumerate(zip(true_labels, pred_labels)):
        if t_true != t_pred:
            original_index = orig_indices[split[row_idx]]

            metadata_val = safe_get_value(sentence_metadata, original_index, 'sentence')
            term_val = safe_get_value(sentence_metadata, original_index, 'term')
            adj_fashion_val = safe_get_value(sentence_metadata, original_index, 'adjectives_fashion')
            adj_char_val = safe_get_value(sentence_metadata, original_index, 'adjectives_char')

            rows.append({
                "index_in_split": split[row_idx],
                "index_in_dataset": original_index,
                "true_label": t_true,
                "predicted_label": t_pred,
                "error_type": (
                    "False Positive" if t_pred == le.classes_[1]
                    else "False Negative"
                ),
                "tokens": features_split[row_idx],
                "num_tokens": len(features_split[row_idx]),
                "metadata": metadata_val,
                "term": term_val,
                "adj_fashion": adj_fashion_val,
                "adj_char": adj_char_val,
            })
    return pd.DataFrame(rows)

```

```python
# Assuming df_errors is your DataFrame
df_errors = df_new
# First, explode the list-like columns into separate rows
df_fashion = df_errors.explode('adj_fashion')
df_term = df_errors.explode('term')
df_char = df_errors.explode('adj_char')

# Function to get most common items for a given error_type and column
def most_common_by_error_type(df, error_type, column, top_n=10):
    filtered = df[df['error_type'] == error_type]
    counts = filtered[column].value_counts()
    return counts.head(top_n)

# Example: top 10 adjectives fashion for False Positives
fp_fashion = most_common_by_error_type(df_fashion, 'False Positive', 'adj_fashion')
print("Top adjectives (fashion) for False Positives:")
print(fp_fashion)

# Example: top 10 adjectives fashion for False Negatives
fn_fashion = most_common_by_error_type(df_fashion, 'False Negative', 'adj_fashion')
print("\nTop adjectives (fashion) for False Negatives:")
print(fn_fashion)

# Similarly for term column
fp_terms = most_common_by_error_type(df_term, 'False Positive', 'term')
fn_terms = most_common_by_error_type(df_term, 'False Negative', 'term')

print("\nTop terms for False Positives:")
print(fp_terms)

print("\nTop terms for False Negatives:")
print(fn_terms)

# And for adj_char column
fp_adj_char = most_common_by_error_type(df_char, 'False Positive', 'adj_char')
fn_adj_char = most_common_by_error_type(df_char, 'False Negative', 'adj_char')

print("\nTop adjectives (character) for False Positives:")
print(fp_adj_char)

print("\nTop adjectives (character) for False Negatives:")
print(fn_adj_char)
```

```python
# Assuming your code above already runs and you have:
# fp_fashion, fn_fashion, fp_terms, fn_terms, fp_adj_char, fn_adj_char

# Function to get unique features between FP and FN
def unique_features(fp_series, fn_series):
    fp_set = set(fp_series.index)
    fn_set = set(fn_series.index)

    unique_fp = fp_set - fn_set
    unique_fn = fn_set - fp_set

    # Return counts for the unique features only
    unique_fp_counts = fp_series[fp_series.index.isin(unique_fp)]
    unique_fn_counts = fn_series[fn_series.index.isin(unique_fn)]

    return unique_fp_counts, unique_fn_counts

# Fashion adjectives
unique_fp_fashion, unique_fn_fashion = unique_features(fp_fashion, fn_fashion)
print("Unique FP fashion adjectives:")
print(unique_fp_fashion)
print("\nUnique FN fashion adjectives:")
print(unique_fn_fashion)

# Terms
unique_fp_terms, unique_fn_terms = unique_features(fp_terms, fn_terms)
print("\nUnique FP terms:")
print(unique_fp_terms)
print("\nUnique FN terms:")
print(unique_fn_terms)

# Character adjectives
unique_fp_adj_char, unique_fn_adj_char = unique_features(fp_adj_char, fn_adj_char)
print("\nUnique FP character adjectives:")
print(unique_fp_adj_char)
print("\nUnique FN character adjectives:")
print(unique_fn_adj_char)
```
