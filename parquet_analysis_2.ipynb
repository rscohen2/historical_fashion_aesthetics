```python

import pandas as pd

# Path to your Parquet file
parquet_file_path_2 = 'fashion_mentions.parquet'

# Load the Parquet file into a Pandas DataFrame
df_2 = pd.read_parquet(parquet_file_path_2)

# Display the first few rows of the DataFrame
print(df_2.head())
```

```python
df_fashion = df_2
df_character = df


df_character['character_id'] = df_character['character_id'].astype(str)
df_fashion['character_id'] = df_fashion['character_id'].astype(str)



# Group fashion terms by character_id and aggregate into a list
fashion_agg = df_fashion.groupby('character_id')['term'].agg(list).reset_index()

# Merge with your character dataframe
df_combined = df_character.merge(fashion_agg, on='character_id', how='left')

# Optional: Replace NaN with empty list if some characters have no fashion data
df_combined['term'] = df_combined['term'].apply(lambda x: x if isinstance(x, list) else [])


```

```python
df_combined
```

```python
from collections import Counter
import pandas as pd

co_occurrence = Counter()

for _, row in df_combined.iterrows():
    for term in list(row['term']):
        for adj in row['adjectives']:
            co_occurrence[(term, adj)] += 1

# Convert to DataFrame
co_occurrence_df = pd.DataFrame([
    {'term': t, 'adjective': a, 'count': c}
    for (t, a), c in co_occurrence.items()
])
```

```python
co_occurrence_df
```

```python
co_occurrence_df = co_occurrence_df[co_occurrence_df['adjective'] != 'sure']
co_occurrence_df = co_occurrence_df[co_occurrence_df['adjective'] != 'glad']
co_occurrence_df = co_occurrence_df[co_occurrence_df['adjective'] != 'afraid']
co_occurrence_df = co_occurrence_df[co_occurrence_df['adjective'] != 'dear']
co_occurrence_df = co_occurrence_df[co_occurrence_df['adjective'] != 'sorry']
co_occurrence_df = co_occurrence_df[co_occurrence_df['adjective'] != 'young']
co_occurrence_df = co_occurrence_df[co_occurrence_df['adjective'] != 'old']
co_occurrence_df = co_occurrence_df[co_occurrence_df['adjective'] != 'able']





```

```python
top_adjectives_per_term = (
    co_occurrence_df
    .sort_values(['term', 'count'], ascending=[True, False])
    .groupby('term')
    .head(2)
)
```
